### 4. Create a class named Rectangle with attributes:
- length
- width

Create a method to calculate and print the area of the rectangle.
```

````python
class Rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def Area(self):
        return self.length*self.width
ob=Rectangle(10,6)
print("Area of Rectangle is",ob.Area())
````

````python
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def Area(self):
        return 22/7*self.radius*self.radius
ob=Circle(7)
print("Area of Circle is",ob.Area())
````

### 6. Create a class named BankAccount with:
- account_holder
- balance

Create methods:
- deposit(amount)
- withdraw(amount)

Display the updated balance after each transaction.
```

````python
class BankAccount:
    def __init__(self,account_holder,balance):
        self.account_holder=account_holder
        self.__balance=balance
    def Deposit(self,amount):
        self.__balance=self.__balance+amount
    def Withdraw(self,amount):
        self.__balance=self.__balance-amount
    def Display(self):
        print("Account Balance:",self.__balance)
ob=BankAccount("Sakshya Chaudhary",5000)
print("Account Holder Name:",ob.account_holder)
ob.Display()
print("Deposit:500")
ob.Deposit(500)
ob.Display()
print("Withdraw:300")
ob.Withdraw(300)
ob.Display()


````

### 7. Create a class named Animal with a method `sound()`.

Create a child class `Dog` that overrides the `sound()` method and prints `"Bark"`.
```

````python
class Animal:
    def sound(self):
        print("Animal Sound")
class Dog(Animal):
    def sound(self):
        print("Bark")
A=Animal()
print("Parent Class:")
A.sound()
print()
print("Child Class:")
D=Dog()
D.sound()
````

### 8. Create a class named Person with attributes:
- name
- age

Create a child class Student with an additional attribute:
- roll_number

Display all the details using an object of the Student class.
```

````python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
class Student(Person):
    def __init__(self, name, age, roll):
        super().__init__(name, age)
        self.roll=roll
ob=Student("Sakshya Chaudhary",20,104)
print("Name:",ob.name)
print("Age:",ob.age)
print("Roll:",ob.roll)
````

### 9. Create a class named Calculator with methods:
- add()
- subtract()
- multiply()
- divide()

Take two numbers as input and perform all operations.
```

````python
class Calculator:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
    def add(self):
        print("Add:",self.num1+self.num2)
    def subtract(self):
        print("Subtract:",self.num1-self.num2)
    def multiply(self):
        print("Multiply:",self.num1*self.num2)
    def divide(self):
        print("Divide:",self.num1/self.num2)
ob=Calculator(20,5)
ob.add()
ob.subtract()
ob.multiply()
ob.divide()
    
````

### 10. Create a class named LibraryBook with attributes:
- book_name
- author
- price

Create a method `display_book_info()` that prints all book details.

Create at least three book objects and display their information.
```

````python
class LibraryBook:
    def __init__(self,book_name,author,price):
        self.book_name=book_name
        self.author=author
        self.price=price
    def display_book_info(self):
        print("Book Name:",self.book_name)
        print("Author:",self.author)
        print("Price:",self.price)
book1=LibraryBook("Python Basics","Rahul Sharma",250)
book2=LibraryBook("Learning C++","Amit Verma",300)
book3=LibraryBook("Data Science Guide","Priya Singh",400)

book1.display_book_info()
print()
book2.display_book_info()
print()
book3.display_book_info()
````
