### 4. Create a class named Rectangle with attributes:
- length
- width

Create a method to calculate and print the area of the rectangle.
```

````python
class Rectangle:
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

shape = Rectangle(15, 5)

print("Area of the rectangle =", shape.area())
````

### 5. Create a class named Circle with an attribute `radius`.

Create a method that calculates and prints the area of the circle.
```

````python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

circle1 = Circle(7)

print("Circle Area:", circle1.area())
````

### 6. Create a class named BankAccount with:
- account_holder
- balance

Create methods:
- deposit(amount)
- withdraw(amount)

Display the updated balance after each transaction.
```

````python
class BankAccount:
    def __init__(self, holder_name, balance):
        self.holder_name = holder_name
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def show_balance(self):
        print("Current Balance:", self.__balance)

acc = BankAccount("Yogesh Sharma", 8000)

print("Account Holder:", acc.holder_name)
acc.show_balance()

print("Deposited: 1000")
acc.deposit(1000)
acc.show_balance()

print("Withdrawn: 500")
acc.withdraw(500)
acc.show_balance()
````

### 8. Create a class named Person with attributes:
- name
- age

Create a child class Student with an additional attribute:
- roll_number

Display all the details using an object of the Student class.
```

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

student1 = Student("Yogesh Sharma", 21, 201)

print("Student Name:", student1.name)
print("Student Age:", student1.age)
print("Roll Number:", student1.roll_no)
````

### 9. Create a class named Calculator with methods:
- add()
- subtract()
- multiply()
- divide()

Take two numbers as input and perform all operations.
```

````python
class SimpleCalculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def addition(self):
        print("Result of Addition =", self.a + self.b)

    def subtraction(self):
        print("Result of Subtraction =", self.a - self.b)

    def multiplication(self):
        print("Result of Multiplication =", self.a * self.b)

    def division(self):
        print("Result of Division =", self.a / self.b)

calc = SimpleCalculator(25, 5)

calc.addition()
calc.subtraction()
calc.multiplication()
calc.division()
````

### 10. Create a class named LibraryBook with attributes:
- book_name

- price

Create a method `display_book_info()` that prints all book details.

Create at least three book objects and display their information.
```

````python
class Book:
    def __init__(self, title, writer, cost):
        self.title = title
        self.writer = writer
        self.cost = cost

    def show_details(self):
        print("Title:", self.title)
        print("Writer:", self.writer)
        print("Cost:", self.cost)

b1 = Book("Python Programming Essentials", "Neha Gupta", 280)
b2 = Book("Advanced C++ Concepts", "Rohan Mehta", 320)
b3 = Book("Data Science Handbook", "Sneha Kapoor", 450)

print("Book 1 Details:")
b1.show_details()

print()
print("Book 2 Details:")
b2.show_details()

print()
print("Book 3 Details:")
b3.show_details()
````
